```python
from transformers import AutoModelForCausalLM
```

```python
model_name = "moonshotai/Kimi-K2-Thinking"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
```

```python
model = AutoModelForCausalLM.from_pretrained("moonshotai/Kimi-K2-Thinking", trust_remote_code=True, dtype="auto")
```

```python
# Move to GPU if available
model = model.eval()

# Example prompt with reasoning enabled
messages = [
    {"role": "user", "content": "Solve this step-by-step: A store offers a 20% discount on a $150 jacket, then takes an additional 15% off the discounted price. What is the final price?"}
]

# Apply chat template and tokenize
input_ids = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    return_tensors="pt"
).to(model.device)

# Generate with reasoning (Kimi K2 Thinking supports step-by-step via special mode)
output_ids = model.generate(
    input_ids,
    max_new_tokens=512,
    do_sample=False,  # Greedy decoding for deterministic reasoning
    temperature=0.0,
    reasoning=True  # This enables the thinking trace (special flag for Kimi-K2-Thinking)
)

# Decode full response
full_response = tokenizer.decode(output_ids[0], skip_special_tokens=False)

# Extract only the assistant's reply
assistant_response = full_response.split("<|assistant|>")[-1].strip()

print("Final Answer:")
print(assistant_response)
```
